# Series
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

```python
A=[1,2,3,4]
data=pd.Series(A)
print(data)
```

if label is not specified then by default it starts with 0.

```python
data[0] # retreving data with help of label
```

```python
data[3] # retreving data with help of label
```

```python
label=['x','y','z','t']
data=pd.Series(A,label)    # creating label for the data
data
```

```python
data['y'] # accessing data with the help of label
```

```python
calories={'day_1':420,'day_2':400,'day_3':450,'day_4':475}
data=pd.Series(calories)    # creating Series with the help of dictionary
data
```

```python
data=pd.Series(calories,index=['day_1','day_2','day_3']) # creating Series for specific label
data
```

```python
del calories,data,A
```

# DataFrame
A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
dict_1={'calories':[420,430,400],'time':[20,30,10]}
data=pd.DataFrame(dict_1) # creating Dataframe for with help of dictionary
data
```

```python
data.loc[0]  # accessing data with label 0.
             # a Series is returned
```

```python
data.loc[[1,2]] # accessing data with label 1 and 2.
                # a DataFrame is returned
```

```python
data=pd.DataFrame(dict_1,index=['day_1','day_2','day_3']) # providing label to the indexes 
data
```

```python
data.loc['day_2'] # accessing data with label name
```

```python
del data,dict_1
```

# Maps

```python
Death_mean=df.Total_Deaths.mean()
df.Total_Deaths.map( lambda p:p-Death_mean) # by this function a new series is generated
```

```python
Total_cases_mean=df.Total_Cases.mean()
```

```python
def fun(row):
    row.Total_Cases=row.Total_Cases-Total_cases_mean
    return row
df.apply(fun,axis='columns') # a new DataFrame is returned where column is updated 
```
